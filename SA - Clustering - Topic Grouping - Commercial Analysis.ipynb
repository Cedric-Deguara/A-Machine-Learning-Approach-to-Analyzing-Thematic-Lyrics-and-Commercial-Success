```python
%pip install pandas
import pandas as pd
%pip install nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import nltk
%pip install numpy
import numpy as np
%pip install matplotlib
%pip install sklearn
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score
%pip install seaborn
import seaborn as sns
%pip install mpl_toolkits
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Read the CSV files (Input Path of df_no_nulls.csv)
VectorizedData = pd.read_csv(r'C:\Users\cedri\Desktop\Datasets and Artifact\notebooks\df_no_nulls.csv')
```

```python
nltk.download('vader_lexicon')
sid = SentimentIntensityAnalyzer()

# Define a function to get the sentiment of a song
def get_sentiment(lyrics):
    
    # Get the sentiment scores
    sentiment_scores = sid.polarity_scores(lyrics)
    
    # Get the compound sentiment score
    compound_score = sentiment_scores['compound']
    
    return compound_score

# Apply the function to your preprocessed lyrics data
VectorizedData['sentiment'] = VectorizedData['lyrics'].apply(get_sentiment)
```

```python
# Iterate over the desired range
for i in np.arange(1, -1.1, -0.1):
    # Calculate the absolute difference between the sentiment scores and the current value
    differences = abs(VectorizedData['sentiment'] - i)
    
    # Get the index of the song with the smallest difference
    index = differences.idxmin()
    
    # Get the song with the smallest difference
    song = VectorizedData.loc[index]
    
    # Print the song and its sentiment score
    print(f"Song: {song['lyrics']}\nSentiment: {song['sentiment']}\n")
```

```python
positive_songs_count = len(VectorizedData[VectorizedData['sentiment'] > 0])
print("Number of positive songs:", positive_songs_count)
```

```python
negative_songs_count = len(VectorizedData[VectorizedData['sentiment'] < 0])
print("Number of negative songs:", negative_songs_count)
```

```python
neutral_songs_count = len(VectorizedData[VectorizedData['sentiment'] == 0])
print("Number of neutral songs:", neutral_songs_count)
```

```python
# Select only 'Sentiment', 'num_words', 'words_per_sec', and 'num_uniq_words' columns
numerical_df = VectorizedData[['sentiment', 'num_words', 'words_per_sec', 'num_uniq_words']]

# Determine the optimal number of clusters
inertia = []
for i in range(1, 11):
    # Perform K-Means clustering with i clusters
    kmeans = KMeans(n_clusters=i, random_state=0).fit(numerical_df)
    # Append the inertia (sum of squared distances to the closest centroid) to the list
    inertia.append(kmeans.inertia_)

# Plot the inertia values
plt.plot(range(1, 11), inertia)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

```python

# Select only 'sentiment', 'num_words', 'words_per_sec', and 'num_uniq_words' columns
numerical_df = VectorizedData[['sentiment', 'num_words', 'words_per_sec', 'num_uniq_words']]

# Fit PCA on the data
pca = PCA().fit(numerical_df)

# Plot the explained variances
features = range(pca.n_components_)
plt.bar(features, pca.explained_variance_ratio_, color='black')
plt.xlabel('PCA features')
plt.ylabel('variance %')
plt.xticks(features)

# Save components to a DataFrame
PCA_components = pd.DataFrame(pca.transform(numerical_df))

plt.show()
```

```python
data = VectorizedData[['sentiment', 'num_words', 'words_per_sec', 'num_uniq_words']]

# Choose the number of clusters (this is something you might need to experiment with)
num_clusters = 3

# Perform K-Means clustering
kmeans = KMeans(n_clusters=num_clusters, random_state=0).fit(data)

# Get the cluster assignments for each song
VectorizedData['cluster'] = kmeans.labels_
```

```python
# Perform PCA to reduce the dimensionality of the data to 2 dimensions
pca = PCA(n_components=2)
pca_result = pca.fit_transform(data)

# Create a scatter plot of the PCA result
plt.figure(figsize=(10, 8))
plt.scatter(pca_result[:, 0], pca_result[:, 1], c=VectorizedData['cluster'], cmap='viridis')
plt.title('Visualization of the clusters')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar(label='Cluster')
plt.show()
```

```python
# Calculate Silhouette Score
silhouette = silhouette_score(data, VectorizedData['cluster'])
print(f'Silhouette Score: {silhouette}')

# Calculate Davies-Bouldin Score
davies_bouldin = davies_bouldin_score(data, VectorizedData['cluster'])
print(f'Davies-Bouldin Score: {davies_bouldin}')

# Calculate Calinski-Harabasz Score
calinski_harabasz = calinski_harabasz_score(data, VectorizedData['cluster'])
print(f'Calinski-Harabasz Score: {calinski_harabasz}')
```

Silhouette Score: 0.514: This score ranges from -1 to 1, with 1 being the best possible score. A score of 0.514 suggests that the clusters are reasonably well separated, but there may still be some overlap.

Davies-Bouldin Score: 0.629: This score ranges from 0 to infinity, with 0 being the best possible score. A score of 0.629 suggests that the clusters are reasonably compact, but there may still be some dispersion.

Calinski-Harabasz Score: 10678: This score is higher when clusters are dense and well separated. There's no upper limit for this score, and higher is better. A score of 10678 is quite high, suggesting that the clusters are well separated.
